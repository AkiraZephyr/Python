```python
import matplotlib.pyplot as plt
import cv2 
img = cv2.imread("C:\\Academics\\Python\\Digital Image Processing\\FullMoon.jpg") 
negative = 255 - img
```

```python
cv2.imshow("Original", img) 
cv2.imshow("Negative", negative) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,5))
# Original Image 
plt.subplot(1,2,1) 
plt.title("Original Image") 
plt.imshow(img) 
plt.axis("off") 

# Equalized Image 
plt.subplot(1,2,2) 
plt.title("Negative Image") 
plt.imshow(negative, cmap='gray') 
plt.axis("off") 
plt.show()
```

```python
img = cv2.imread("C:\\Academics\\Python\\Digital Image Processing\\ILena.jpg")  # Replace with your image path 
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
negative_color = 255 - img
cv2.imshow("Original Color", img) 
cv2.imshow("Negative Color", negative_color)
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
plt.figure(figsize=(10,5))
# Original Image 
plt.subplot(1,2,1) 
plt.title("Original Image") 
plt.imshow(img_rgb) 
plt.axis("off") 
negative_color = 255 - img_rgb
# Equalized Image 
plt.subplot(1,2,2) 
plt.title("Negative Image") 
plt.imshow(negative_color) 
plt.axis("off") 
plt.show()
```
