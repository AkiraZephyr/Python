```python
import cv2 
# Load image in color (or grayscale for simpler visualization) 
img = cv2.imread("C:\\Academics\\Python\\Digital Image Processing\\Satellite.png", cv2.IMREAD_COLOR) 
# Check if image loaded 
if img is None: 
	raise FileNotFoundError("Satellite.png not found")
```

```python
# Resize to 800x600 for easier display 
resized_img = cv2.resize(img, (800, 600))
```

```python
cv2.imshow("Satellite Image", resized_img) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
#Convert to grayscale
gray = cv2.cvtColor(resized_img, cv2.COLOR_BGR2GRAY) 
cv2.imshow("Grayscale Satellite", gray) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
#Contrast Enhancement using Histogram Equalization
# Apply Histogram Equalization 
import numpy as np 
equalized = cv2.equalizeHist(gray) 
cv2.imshow("Enhanced Contrast", equalized) 
cv2.waitKey(0) 
cv2.destroyAllWindows() 
```

```python
#Zoom/Visualize Different Regions 
# Example: Crop a region 
roi = equalized[100:400, 200:600]  # [y1:y2, x1:x2] 
cv2.imshow("Region of Interest", roi) 
cv2.waitKey(0)
cv2.destroyAllWindows() 
```

```python
#save the images
cv2.imwrite("satellite_enhanced.jpg", equalized) 
cv2.imwrite("satellite_roi.jpg", roi) 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(12, 6))

# Show original resized image
plt.subplot(1, 4, 1)
plt.imshow(cv2.cvtColor(resized_img, cv2.COLOR_BGR2RGB))
plt.title('Resized')
plt.axis('off')

# Show grayscale image
plt.subplot(1, 4, 2)
plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.axis('off')

# Show histogram equalized image
plt.subplot(1, 4, 3)
plt.imshow(equalized, cmap='gray')
plt.title('Equalized')
plt.axis('off')

# Show ROI
plt.subplot(1, 4, 4)
plt.imshow(roi, cmap='gray')
plt.title('ROI')
plt.axis('off')

plt.tight_layout()
plt.show()
```
