```python
import cv2 
img = cv2.imread("C:\\Academics\\Python\\Digital Image Processing\\ILena.jpg") 
smoothed = cv2.blur(img, (5,5)) 
```

```python
cv2.imshow("Original", img) 
cv2.imshow("Smoothed", smoothed) 
cv2.waitKey(0) 
cv2.destroyAllWindows()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))

# Show original image (convert BGR to RGB)
plt.subplot(1, 2, 1)
plt.imshow(img[..., ::-1])
plt.title('Original Image')
plt.axis('off')

# Show smoothed image (convert BGR to RGB)
plt.subplot(1, 2, 2)
plt.imshow(smoothed[..., ::-1])
plt.title('Smoothed Image')
plt.axis('off')

plt.tight_layout(w_pad=2)
plt.show()
```
